1.Define the role of a Data Analyst in your own words. What value do they bring to an organization?

  - The role of a Data Analyst is to collect, clean, organize, and analyze data so that a company can make better decisions. A data analyst looks at raw data, finds patterns, creates reports or dashboards, and helps explain what the data means in a simple way for the team or management.
  
  -  the value a data analyst bring to an organization is that they turn data into useful insights. Instead of guessing, a company can rely on data to understand what is working, what is not, and where improvements are needed. This helps the organization save time, reduce mistakes, improve performance, and make smarter business decisions. In short, a data analyst helps a company grow by using data to support every important decision.

2.List three tools commonly used by Data Analysts and explain what each is mainly used for.

  -  Microsoft Excel=Excel is used for organizing data, doing basic calculations, creating charts, and performing simple analysis. It is one of the easiest tools for beginners and is widely used in companies for quick data tasks.
  - SQL (Structured Query Language)=SQL is used to retrieve, filter, and manage data stored in databases. Data analysts use SQL to find specific information from large datasets, such as customer details, sales numbers, or performance reports.
  - Tableau (or Power BI)=These are data visualization tools used to create dashboards and interactive charts. They help analysts present data in a clear and visual way so that managers can understand insights quickly and make decisions.

3.Write the end-to-end analytics workflow in the correct order and briefly explain each step.

  - Define the Problem=The first step is to clearly understand what question needs to be answered or what business problem needs to be solved. This ensures the analysis has a clear purpose.
  - Collect the Data=Gather the required data from different sources such as databases, files, surveys, or company systems. Without the right data, analysis cannot begin.
  - Clean and Prepare the Data=Data often contains errors, missing values, or duplicate entries. In this step, the analyst cleans the data and organizes it into a usable format. This is one of the most important steps.
  - Explore and Analyze the Data=Here, the analyst looks for patterns, trends, and relationships using statistical methods, charts, or queries. This helps in understanding what the data reveals.
  - Build Insights or Models=Based on the analysis, insights are generated or simple models may be created to answer the problem. This step connects data findings to business needs.
  - Visualize the Results=Dashboards, charts, and graphs are created to clearly present the key insights so that non-technical stakeholders can easily understand the outcome.
  - Communicate Findings & Recommend Actions=Finally, the analyst explains the results to the team or management and gives suggestions on what actions should be taken based on the insights.
  
4.What is Prompt Engineering, and explain any two types of prompts used in Generative AI?

  - A Prompt Engineer is a person who designs and writes effective prompts (instructions or questions) for generative AI tools like ChatGPT, Midjourney, or other AI models.Their job is to understand how the AI works and create prompts that give accurate, useful, and high-quality outputs.In simple words, a prompt engineer guides the AI by asking the right questions in the right way.
  - Instruction Prompt
This type of prompt gives clear instructions about what the AI should do.
Example: “Summarize this paragraph in simple language.”
It tells the AI exactly what task to perform.

  - Context Prompt (or Information Prompt)
This prompt provides background information or context so the AI can give a better answer.

5.What are the key differences between a Business Analyst and a Data Analyst in terms of roles,responsibilities, and focus?

  - Focus Area:Business Analyst (BA):
Focuses on understanding business needs, improving processes, and helping the company make strategic decisions. BA works closely with stakeholders to define business requirements.

Data Analyst (DA):
Focuses on working with data—collecting, cleaning, analyzing, and visualizing data to find patterns and insights that support decision-making.

  - Responsibilities

Business Analyst:Gather and understand business requirements.
Communicate with departments to identify problems and opportunities.
Recommend process improvements and business solutions.
Create documentation like BRDs (Business Requirement Documents) and workflows.

Data Analyst:
Collect and clean data from various sources.
Perform data analysis using tools like SQL, Excel, Python, etc.
Create charts, dashboards, and reports.
Interpret data to provide insights and answer specific business questions.

   - Role Perspective

Business Analyst:
Looks at business processes and how to improve them. Their perspective is more strategic and operational.

Data Analyst:
Looks at data and what it reveals. Their perspective is more technical and analytical.

8.Explain the evolution of analytics, highlighting the different stages from Descriptive to Generative AI.

  - Descriptive Analytics (What happened?)

Descriptive analytics focuses on summarizing past data to understand what has already happened.

  - Diagnostic Analytics (Why did it happen?)

Diagnostic analytics goes deeper to find the reasons behind past outcomes.

  - Predictive Analytics (What will happen?)

Predictive analytics uses statistical models and machine learning to forecast future events.

  -  Prescriptive Analytics (What should we do?)

Prescriptive analytics suggests actions or decisions to achieve the best outcome.

  - Generative AI: Create insights automatically and simplify analytics
